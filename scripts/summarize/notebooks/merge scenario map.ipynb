```python
import pandas as pd
import os
```

```python
scenario_file_locs = {'2014':r'L:\vision2050\soundcast\non_integrated\soundcast_2014_Vision','2050 No Action':r'L:\vision2050\soundcast\non_integrated\2050\stc_no_action',
                      '2050 Distributed Growth':r'L:\vision2050\soundcast\non_integrated\2050\dug'}
output_file = r'L:\vision2050\soundcast\non_integrated\soundcast_2014_Vision\outputs\merged_maps.csv'
compare = {'2014':'2050 No Action', '2050 No Action':'2050 Distributed Growth'}
```

```python
i=0
for name, path in scenario_file_locs.iteritems():
    if i == 0:
        scenario_results = pd.read_csv(os.path.join(path,r'outputs\Vision2050_maps.csv' ))
        scenario_results['name'] = name
        first_zones = scenario_results['hhtaz']
        
    else: 
        scenario = pd.read_csv(os.path.join(path,r'outputs\Vision2050_maps.csv' ))
        scenario['name'] = name
        scenario = scenario[scenario['hhtaz'].isin(first_zones.tolist())]
        scenario_results = pd.concat([scenario, scenario_results])
    i=i+1
```

```python
# now calculate differences
for base, alt in compare.iteritems():
    scenario_results_base = scenario_results[scenario_results['name']==base]
    first_zones = scenario_results_base['hhtaz']
    #scenario_results_base.set_index('hhtaz', inplace = True)
    scenario_results_alt = scenario_results[scenario_results['name']==alt]
    #scenario_results_alt.set_index('hhtaz', inplace = True)
    scenario_results_alt= scenario_results_alt[scenario_results_alt['hhtaz'].isin(first_zones.tolist())]
    second_zones = scenario_results_alt['hhtaz']
    scenario_results_base = scenario_results_base[scenario_results_base['hhtaz'].isin(second_zones.tolist())]
    
    scenario_results_base= scenario_results_base.set_index('hhtaz')
    scenario_results_alt= scenario_results_alt.set_index('hhtaz')
    compare_scenarios = scenario_results_alt[['annual hours of delay', 'daily vehicle miles', 'percent of people walking or biking']].sub(scenario_results_base[['annual hours of delay', 'daily vehicle miles', 'percent of people walking or biking']])
    compare_scenarios.reset_index(inplace =True)
    compare_scenarios['name'] = alt + '-'+ base
    scenario_results = pd.concat([compare_scenarios, scenario_results])
```

```python
scenario_results.to_csv(output_file)
```
