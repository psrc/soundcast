The trip mode choice models explanatory variables include household and person variables, level-of-service between the trip origin and destination according to the time period for the tour leg, urban form variables, and alternative-specific constants segmented by tour mode.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import os, sys

module_path = os.path.abspath(os.path.join('../../../..')) # or the path to your source code
sys.path.insert(0, module_path)
import util
```

```python
# read data
person = util.get_person_data(summary_config)
hh = util.get_hh_data(summary_config)
trip = util.get_trip_data(summary_config)
tour = util.get_tour_data(summary_config)
```

```python
per_data = person.merge(hh[['household_id','auto_ownership','auto_ownership_4+','log_hh_1','log_hh_1_group','source']],
                          how='left', on=['household_id','source']) # get auto ownership from hh data

trip_data = trip.\
    merge(tour[['tour_id', 'tour_mode', 'tour_mode_transit_agg', 'tour_type','source']], how = 'left', on=['tour_id', 'source']).\
    merge(per_data, how = 'left', on=['person_id', 'household_id', 'source'])
```

```python
util.plot_share_barchart(trip_data, 'trip_weight', 'trip_mode_transit_agg', title="Trip mode choice: all modes")
```

```python
util.plot_share_barchart(trip_data, 'trip_weight', 'trip_mode_transit_agg', title="Trip mode choice: transit modes")
```

## Trip mode choice by segment

```python
def trip_mode_choice_seg(segment_col, facet_col_wrap=3, height=[1200, 800]):
    """
    faceted bar chart showing trip mode share by segment
    """
    util.plot_share_facetbar(trip_data, 'trip_weight', share_col="trip_mode_transit_agg", title=f'Trip mode choice by {segment_col}', facet_col=segment_col, facet_col_wrap=facet_col_wrap, height=height[0])
    util.plot_share_facetbar(trip_data, 'trip_weight', share_col="trip_mode_transit_only", title=f'Trip mode choice by {segment_col}: transit only', facet_col=segment_col, facet_col_wrap=facet_col_wrap, height=height[1])
```

```python
trip_mode_choice_seg(segment_col='purpose')
```

```python
trip_mode_choice_seg(segment_col='ptype_label', facet_col_wrap=2, height=[1000, 700])
```

```python
trip_mode_choice_seg(segment_col='log_hh_1_group', height=[600, 500])
```

```python
trip_mode_choice_seg(segment_col='auto_ownership_4+', height=[600, 500])
```
