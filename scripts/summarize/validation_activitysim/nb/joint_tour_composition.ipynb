```python
import pandas as pd
import numpy as np
import plotly.express as px
import os, sys

module_path = os.path.abspath(os.path.join('../../../..')) # or the path to your source code
sys.path.insert(0, module_path)
import util
```

```python
# read data
hh = util.get_hh_data(summary_config, uncloned=True)
person = util.get_person_data(summary_config, uncloned=True)
tour = util.get_tour_data(summary_config)
```

This model predicts the makeup of the travel party (adults, children, or mixed).

```python
# all joint tours
num_hh_joint_tours = tour.loc[tour['tour_category'] == "joint"].groupby(['source','household_id']).size().to_frame(). \
    rename(columns={0: 'num_hh_joint_tours'})

# all households with joint tours
hh_data = pd.merge(hh, num_hh_joint_tours, on=['source','household_id'], how='left')
hh_data['num_hh_joint_tours'] = hh_data['num_hh_joint_tours'].fillna(0)
joint_hh_data = hh_data.loc[hh_data['num_hh_joint_tours']>0]

per_data = person.\
    merge(hh_data[['household_id','income_group','num_drivers','auto_ownership_4+','auto_available_driver','hhsize_4+','num_hh_joint_tours','hh_weight','source']],
            how='left', on=['household_id','source'])

tour_data = tour.merge(per_data, how='left', on=['person_id','household_id','source'])
joint_tour_data = tour_data.loc[tour_data['tour_category'] == "joint"].copy()
```

- joint tour count

```python
# total number of joint tours by source
df_1 = joint_tour_data.groupby('source')['tour_weight'].sum().reset_index()
df_1 = df_1.rename(columns={"tour_weight": "joint tour count"})
# total number of tours by source
df_2 = tour.groupby('source')['tour_weight'].sum().reset_index()
df_2 = df_2.rename(columns={"tour_weight": "tour count"})

df_joint_tour_count = df_1.merge(df_2, how="left", on='source')
df_joint_tour_count['joint_percent'] = df_joint_tour_count['joint tour count']/df_joint_tour_count['tour count']
df_joint_tour_count['tour count'] = df_joint_tour_count['tour count'].astype(float).map(lambda n: '{:.2f}'.format(n))
df_joint_tour_count['joint tour count'] = df_joint_tour_count['joint tour count'].astype(float).map(lambda n: '{:.2f}'.format(n))
df_joint_tour_count['joint_percent'] = df_joint_tour_count['joint_percent'].astype(float).map(lambda n: '{:.2%}'.format(n))
df_joint_tour_count
```
