## trip distance

```python
util.plot_share_barchart(trip_data, 'trip_weight', 'trip_distance_bin', title="share of trip by distance bin")
```

```python
df_plot = trip_data.loc[trip_data['trip_distance_bin_60mi'].notna()].groupby(['source','trip_distance_bin_60mi'])['trip_weight'].sum().reset_index()
df_plot['percentage'] = df_plot.groupby(['source'], group_keys=False)['trip_weight']. \
    apply(lambda x: x / float(x.sum()))

fig2 = px.line(df_plot, x='trip_distance_bin_60mi', y="percentage", color="source", title="trip distance")
fig2.for_each_annotation(lambda a: a.update(text = a.text.split("=")[-1]))
fig2.update_layout(height=400, width=700, font=dict(size=11), yaxis=dict(tickformat=".1%"))
fig2.show()
```
