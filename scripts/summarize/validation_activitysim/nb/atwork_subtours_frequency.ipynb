The at-work subtour frequency model selects the number of at-work subtours made for each work tour.

- work tours Alternatives: none, 1 eating out tour, 1 business tour, 1 maintenance tour, 2 business tours, 1 eating out tour + 1 business tour
- Explanatory variables include employment status, income, auto ownership, the frequency of other tours, characteristics of the parent work tour, and characteristics of the workplace zone.


```python
import pandas as pd
import polars as pl
import plotly.express as px
import os, sys

module_path = os.path.abspath(os.path.join('../../../..')) # or the path to your source code
sys.path.insert(0, module_path)
import util
```

```python
# read data
hh = util.get_hh_data(summary_config, uncloned=True)
person = util.get_person_data(summary_config, uncloned=True)
tour = util.get_tour_data(summary_config)
```

```python
per_data = person.merge(hh[['household_id','income_group','auto_ownership_4+','log_hh_1','log_hh_1_group',
                            'auto_available_driver','auto_available_worker','source']],
                          how='left', on=['household_id','source']) # get auto ownership from hh data

tour_data = tour.merge(per_data, how='left', on=['person_id','household_id','source'])
```

```python
atwork_subtour = tour_data.loc[tour_data['tour_category']=="atwork"].copy()
# select all work tours that could make atwork subtours
work_tour = tour_data.loc[tour_data['tour_type']=="work"].copy()

# calculate atwork_subtour_frequency

# count number of subtours by type
tour_type_count = atwork_subtour.groupby(['source',"parent_tour_id","tour_type"],observed=True, as_index=False).size()
df_wide = tour_type_count.pivot(index=['source','parent_tour_id'], columns='tour_type', values='size').reset_index()
df_wide = df_wide.fillna(0)
df_wide['subtour_count'] = df_wide['eat'] + df_wide['business'] + df_wide['maint']

# assign atwork_subtour_frequency for parent tour id
df_wide['atwork_subtour_frequency_reassign'] = ""
df_wide.loc[df_wide['maint']>=1.0,'atwork_subtour_frequency_reassign'] = 'maint'
df_wide.loc[df_wide['eat']>=1.0,'atwork_subtour_frequency_reassign'] = 'eat'
df_wide.loc[df_wide['business']==1.0,'atwork_subtour_frequency_reassign'] = 'business1'
df_wide.loc[df_wide['business']>1.0,'atwork_subtour_frequency_reassign'] = 'business2'
df_wide.loc[(df_wide['eat']==1.0) & (df_wide['business']==1.0),'atwork_subtour_frequency_reassign'] = 'eat_business'

# merge to parent work tour (replace missing only: survey records)
work_tour = work_tour.merge(df_wide, left_on=['source','tour_id'],right_on=['source','parent_tour_id'], how='left')
work_tour.loc[work_tour['atwork_subtour_frequency'].isna(),'atwork_subtour_frequency'] = work_tour.loc[work_tour['atwork_subtour_frequency'].isna(),'atwork_subtour_frequency_reassign']
work_tour['atwork_subtour_frequency'] = work_tour['atwork_subtour_frequency'].fillna('no_subtours')

# max number of subtours at 2
work_tour['subtour_count'] = work_tour['subtour_count'].fillna(0)
work_tour.loc[work_tour['subtour_count']>2,'subtour_count'] = 2
```

```python
util.plot_share_barchart(work_tour, 'tour_weight', 'atwork_subtour_frequency', title="atwork subtours frequency")
```
