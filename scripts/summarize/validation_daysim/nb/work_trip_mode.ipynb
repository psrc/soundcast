```python
import plotly.express as px
import polars as pl
import os, sys

module_path = os.path.abspath(os.path.join('../../../..')) # or the path to your source code
sys.path.insert(0, module_path)
import util
```

```python
hh = util.get_hh_data(summary_config)
person = util.get_person_data(summary_config)
trip = util.get_trip_data(summary_config)
parcel_geog = util.get_parcel_geog(input_config, summary_config)
```

```python
# add column aliases for segment plots
hh = hh.join(parcel_geog, left_on='hhparcel', right_on='ParcelID', how='left')
hh = hh.with_columns([
    pl.col('hhvehs_4+').alias('auto_ownership_4+'),
    pl.col('auto_available_hhwkrs').alias('auto_available_worker'),
    pl.col('auto_available_drivers').alias('auto_available_driver'),
])

person = person.join(hh, how='left', on=['hhno', 'source'])
trip = trip.join(person, how='left', on=['pno', 'hhno', 'source'])
```

```python
wk_trip = trip.filter(pl.col('dpurp') == 1)
```

```python
# Load ACS data
df_acs = util.read_sqlite_db(input_config, summary_config, "SELECT * FROM acs_commute_mode_by_home_county")

df_acs = df_acs.rename({'share': 'percentage'}).with_columns([
    pl.lit('ACS').alias('source'),
    pl.lit(None).cast(pl.Float64).alias('trexpfac')
])

mode_cat_acs = {'Walk': "1: walk",
                'Bike': "2: bike",
                'SOV': "3: sov",
                'HOV2': "4: hov 2",
                'HOV3+': "5: hov 3+",
                'Transit': "6: transit",
                'TNC': "9: tnc"}
df_acs = df_acs.with_columns(pl.col('mode').replace_strict(mode_cat_acs, default=None).alias('mode_label'))

df_acs_region = df_acs.filter(pl.col('county') == "Region").select(['source', 'mode_label', 'trexpfac', 'percentage'])
```

```python
df_plot = (
    wk_trip.group_by(['source', 'mode_label'])
    .agg([
        pl.col('trexpfac').sum().alias('trexpfac'),
        pl.col('trexpfac').count().alias('sample count')
    ])
    .with_columns((pl.col('trexpfac') / pl.col('trexpfac').sum().over(['source'])).alias('percentage'))
)
df_plot = pl.concat([df_plot, df_acs_region], how='diagonal')

fig = px.bar(df_plot.sort(['source']).to_pandas(), x="mode_label", y="percentage",
             color="source", barmode="group",
             category_orders={'source': ['model', 'survey', 'ACS']},
             hover_data=['sample count'],
             title="work trip mode")
fig.update_layout(height=400, width=700, font=dict(size=11),
                  xaxis=dict(dtick=1, categoryorder='category ascending'),
                  yaxis=dict(tickformat=".2%"))
fig.show()
```

### mode choice by segment

```python
df_plot = (
    wk_trip.group_by(['source', 'CountyName', 'mode_label'])
    .agg([
        pl.col('trexpfac').sum().alias('trexpfac'),
        pl.col('trexpfac').count().alias('sample count')
    ])
    .with_columns((pl.col('trexpfac') / pl.col('trexpfac').sum().over(['source', 'CountyName'])).alias('percentage'))
)

df_acs_county = (
    df_acs.filter(pl.col('county') != "Region")
    .select(['source', 'county', 'mode_label', 'trexpfac', 'percentage'])
    .rename({'county': 'CountyName'})
)
df_plot = pl.concat([df_plot, df_acs_county], how='diagonal').filter(pl.col('CountyName') != "Outside Region")

fig = px.bar(df_plot.sort(['mode_label'], descending=True).to_pandas(),
             x="percentage", y="mode_label", color="source", barmode="group",
             facet_col='CountyName', facet_col_wrap=2, orientation='h',
             hover_data=['sample count'],
             category_orders={"source": ["model", "survey", "ACS"]},
             title="work trip mode choice by home county")
fig.update_layout(width=700, height=550)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.for_each_xaxis(lambda a: a.update(tickformat=".1%"))
fig.show()
```

```python
def plot_mode_choice(df: pl.DataFrame, grp_var: str, order_list: dict, title_name: str, n_nol: int, height=400, width=800):
    df_plot = (
        df.group_by(['source', grp_var, 'mode_label'])
        .agg([
            pl.col('trexpfac').sum().alias('trexpfac'),
            pl.col('trexpfac').count().alias('sample count')
        ])
        .with_columns(
            (pl.col('trexpfac') / pl.col('trexpfac').sum().over(['source', grp_var])).alias('percentage')
        )
    )

    fig = px.bar(df_plot.sort(['source', 'mode_label']).to_pandas(),
                 x="percentage", y="mode_label", color="source", barmode="group",
                 facet_col=grp_var, facet_col_wrap=n_nol, orientation='h',
                 hover_data=['sample count'],
                 category_orders=order_list,
                 title="work trip mode choice by " + title_name)
    fig.update_layout(height=height, width=width)
    fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig.for_each_xaxis(lambda a: a.update(tickformat=".1%"))
    fig.show()
```
