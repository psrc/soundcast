Person Types

```python
# add person type labels
ptype_cat = {1: "1: full time worker",
             2: "2: part time worker",
             3: "3: non-worker age 65+",
             4: "4: other non-working adult",
             5: "5: university student",
             6: "6: grade school student/child age 16+",
             7: "7: child age 5-15",
             8: "8: child age 0-4"}
person = person.with_columns(pl.col('pptyp').replace_strict(ptype_cat, default=None).alias('pptyp_label'))
```

```python
df_plot = (
    person.group_by(['source', 'pptyp_label'])
    .agg([
        pl.col('psexpfac').sum().alias('psexpfac'),
        pl.col('psexpfac').count().alias('sample count')
    ])
    .with_columns((pl.col('psexpfac') / pl.col('psexpfac').sum().over(['source'])).alias('percentage'))
)

fig = px.bar(df_plot.sort(['source']).to_pandas(), x="pptyp_label", y="percentage", color="source",
             hover_data=['sample count'],
             barmode="group", title="person type")
fig.update_layout(height=400, width=700, font=dict(size=11),
                  yaxis=dict(tickformat=".2%"))
fig.show()
```
