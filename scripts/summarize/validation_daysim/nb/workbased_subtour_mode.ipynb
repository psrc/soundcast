```python
hh = util.get_hh_data(summary_config)
person = util.get_person_data(summary_config)
tour = util.get_tour_data(summary_config)
parcel_geog = util.get_parcel_geog(input_config, summary_config)
```

```python
# join work parcel geography to person for work county
person = person.join(parcel_geog, how="left", left_on='pwpcl', right_on='ParcelID')

# add column aliases for segment plots
hh = hh.with_columns([
    pl.col('hhvehs_4+').alias('auto_ownership_4+'),
    pl.col('auto_available_drivers').alias('auto_available_driver'),
])

person = person.join(hh, how='left', on=['hhno', 'source'])
tour = tour.join(person, how='left', on=['pno', 'hhno', 'source'])
```

```python
wk_base_subtour = tour.filter(pl.col('parent') != 0)
wk_tour = tour.filter((pl.col('parent') == 0) & (pl.col('pdpurp') == 1))
```

```python
test = wk_tour.filter(pl.col('subtrs') != 11)
df_plot = (
    test.group_by(['source', 'subtrs'])
    .agg([
        pl.col('toexpfac').sum().alias('toexpfac'),
        pl.col('toexpfac').count().alias('sample count')
    ])
    .with_columns((pl.col('toexpfac') / pl.col('toexpfac').sum().over(['source'])).alias('percentage'))
)

fig = px.bar(df_plot.sort(['source']).to_pandas(), x="subtrs", y="percentage", color="source",
             barmode="group", hover_data=['sample count'], title="number of workbased subtours (only work tours)")
fig.update_layout(height=400, width=700, font=dict(size=11),
                  xaxis=dict(dtick=1, categoryorder='category ascending'),
                  yaxis=dict(tickformat=".2%"))
fig.show()
```

### mode choice by segment

```python
def plot_mode_choice(df: pl.DataFrame, grp_var: str, order_list: dict, title_name: str, n_nol: int, height=400, width=800):
    df_plot = (
        df.group_by(['source', grp_var, 'tmodetp_label'])
        .agg([
            pl.col('toexpfac').sum().alias('toexpfac'),
            pl.col('toexpfac').count().alias('sample count')
        ])
        .with_columns(
            (pl.col('toexpfac') / pl.col('toexpfac').sum().over(['source', grp_var])).alias('percentage')
        )
    )

    fig = px.bar(df_plot.sort(['source', 'tmodetp_label']).to_pandas(),
                 x="percentage", y="tmodetp_label", color="source", barmode="group",
                 facet_col=grp_var, facet_col_wrap=n_nol, orientation='h',
                 hover_data=['sample count'],
                 category_orders=order_list,
                 title="work tour mode choice by " + title_name)
    fig.update_layout(height=height, width=width)
    fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig.for_each_xaxis(lambda a: a.update(tickformat=".1%"))
    fig.show()
```

```python
order_mode = ["1: walk", "2: bike", "3: sov", "4: hov 2", "5: hov 3+", "6: walk to transit",
              "7: park-and-ride", "8: school bus", "9: tnc"]
incl_county = ["King", "Kitsap", "Pierce", "Snohomish"]

plot_mode_choice(wk_base_subtour.filter(pl.col("CountyName").is_in(incl_county)), "CountyName",
                 {"CountyName": ["King", "Kitsap", "Pierce", "Snohomish"],
                  "tmodetp_label": order_mode},
                 "work county",
                 2, 600)
```

```python
plot_mode_choice(wk_base_subtour.filter(pl.col('auto_ownership_4+') != "-1"), "auto_ownership_4+",
                 {"auto_ownership_4+": ["0", "1", "2", "3", "4+"],
                  "tmodetp_label": order_mode},
                 "auto ownership", 2, 800)
```

```python
plot_mode_choice(wk_base_subtour.filter(pl.col('auto_available_driver') != "no driver"), "auto_available_driver",
                 {"auto_available_driver": ["no car", "cars fewer than drivers", "enough cars"],
                  "tmodetp_label": order_mode},
                 "auto availability (driver, showing only households with at least one driver)", 3, 500, 1000)
```
