```python
import plotly.express as px
import polars as pl
import os, sys

module_path = os.path.abspath(os.path.join('../../../..')) # or the path to your source code
sys.path.insert(0, module_path)
import util
```

```python
trip = util.get_trip_data(summary_config)
```

```python
max_bin = 60
bin_size = 2

trip = trip.with_columns([
    pl.when(pl.col('travdist') < 1).then(pl.lit('0-1'))
      .when(pl.col('travdist') < 3.5).then(pl.lit('1-3.5'))
      .when(pl.col('travdist') < 10).then(pl.lit('3.5-10'))
      .when(pl.col('travdist') >= 10).then(pl.lit('10+'))
      .otherwise(None)
      .alias('distance_bin'),
    pl.when((pl.col('travdist') >= 0) & (pl.col('travdist') <= max_bin))
      .then(pl.col('travdist').truediv(bin_size).floor().mul(bin_size).cast(pl.Int64).cast(pl.Utf8))
      .otherwise(None)
      .alias('bin_60mi')
])
```

```python
def plot_bins(df: pl.DataFrame, filter_expr=None, title_name=None):

    if filter_expr is not None:
        df = df.filter(filter_expr)

    df_plot = (
        df.filter(pl.col('distance_bin').is_not_null())
        .group_by(['source', 'distance_bin'])
        .agg([
            pl.col('trexpfac').sum().alias('trexpfac'),
            pl.col('trexpfac').count().alias('sample count')
        ])
        .with_columns((100 * pl.col('trexpfac') / pl.col('trexpfac').sum().over(['source'])).alias('percentage'))
    )

    if title_name:
        chart_title = "Trip distance by purpose: " + title_name
    else:
        chart_title = "Trip distance"

    fig = px.bar(df_plot.to_pandas(), x="distance_bin", y="percentage", color="source", barmode="group",
                 facet_col_wrap=2, hover_data=['sample count'],
                 title=chart_title)
    fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig.update_layout(height=400, width=700, font=dict(size=11))
    fig.show()

plot_bins(trip)
```
