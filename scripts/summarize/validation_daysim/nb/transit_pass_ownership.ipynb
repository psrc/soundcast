```python
import plotly.express as px
import polars as pl
import os, sys

module_path = os.path.abspath(os.path.join('../../../..')) # or the path to your source code
sys.path.insert(0, module_path)
import util
```

```python
hh = util.get_hh_data(summary_config, quantile_groups=['hhincome', 'emptot_1', 'hh_1'])
person = util.get_person_data(summary_config)
parcel_geog = util.get_parcel_geog(input_config, summary_config)
```

```python
# Create column aliases for group columns
hh = hh.with_columns([
    pl.col('hhincome_4group').alias('hhincome_group'),
    pl.col('hh_1_4group').alias('hh_density_group'),
    pl.col('emptot_1_4group').alias('emp_density_group'),
])
```

```python
# Group ages
survey_ages = (
    person.filter(pl.col('source') == 'survey')
    .select(pl.col('pagey').cast(pl.Int64))
    .to_series().unique().sort().to_list()
)
survey_ages = survey_ages + [999]

# Create labels
person = person.with_columns(pl.col('pagey').cut(survey_ages).alias('age'))
```

```python
person = person.join(hh, on=['hhno', 'source'], how='left')
```

```python
person = person.join(parcel_geog, left_on='hhparcel', right_on='ParcelID', how='left')
```

```python
df = (
    person.group_by(['source', 'ptpass'])
    .agg(pl.col('psexpfac').sum().alias('psexpfac'))
    .pivot(values='psexpfac', index='source', on='ptpass')
    .with_columns([
        (pl.col('0') / pl.col('0').fill_null(0).add(pl.col('1').fill_null(0))).alias('no pass'),
        (pl.col('1') / pl.col('0').fill_null(0).add(pl.col('1').fill_null(0))).alias('% owns transit pass')
    ])
)
df.select(['source', '% owns transit pass'])
```

```python
# auto ownership in Income groups
def plot_transit_pass(df: pl.DataFrame, var: str, title_cat: str, sub_name: str):

    df_plot = (
        df.group_by(['source', var, 'ptpass'])
        .agg([
            pl.col('psexpfac').sum().alias('psexpfac'),
            pl.col('psexpfac').count().alias('sample count')
        ])
        .with_columns(
            (pl.col('psexpfac') / pl.col('psexpfac').sum().over(['source', var])).alias('percentage')
        )
        .filter(pl.col('ptpass') == 1)
    )

    try:
        df_plot = df_plot.with_columns(
            pl.col(var).cast(pl.Utf8).str.extract(r'(\d+)', 1).cast(pl.Int64).alias('label_numeric')
        ).sort('label_numeric')
    except:
        pass

    fig = px.bar(df_plot.to_pandas(), x=var, y="percentage", color="source",
                 barmode="group",
                 hover_data=['sample count'],
                 title="Transit Pass Ownership by " + title_cat)

    fig.update_layout(xaxis_title=sub_name)

    fig.show()
```

## Pass Ownership by Home Location

```python
plot_transit_pass(person, 'CountyName', 'Home County', 'County')
```

```python
plot_transit_pass(person, 'rg_proposed', 'Regional Geography', 'Geog')
```

```python
plot_transit_pass(person, 'pptyp_label', 'Person Type', 'person type')
```
