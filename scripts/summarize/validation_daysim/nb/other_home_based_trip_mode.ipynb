```python
import plotly.express as px
import polars as pl
import os, sys

module_path = os.path.abspath(os.path.join('../../../..')) # or the path to your source code
sys.path.insert(0, module_path)
import util
```

```python
hh = util.get_hh_data(summary_config)
person = util.get_person_data(summary_config)
trip = util.get_trip_data(summary_config)
```

```python
df_trip = trip.clone()
df_hh = hh.clone()
df_person = person.clone()

# Add column for (potential) drivers adults (all hh members 16 and above)
df_hh = df_hh.with_columns((
    pl.col('hhsize') - pl.col('hh515') - pl.col('hhcu5')
).alias('drivers'))
# auto_ownership with 4+
df_hh = df_hh.with_columns(
    pl.when(pl.col('hhvehs') >= 4.0).then(pl.lit("4+")).otherwise(pl.col('hhvehs').cast(pl.Utf8)).alias('auto_ownership_4+')
)
# hhsize with 4+
df_hh = df_hh.with_columns(
    pl.when(pl.col('hhsize') >= 4.0).then(pl.lit("4+")).otherwise(pl.col('hhsize').cast(pl.Utf8)).alias('hhsize_4+')
)
# auto availability
df_hh = df_hh.with_columns([
    (pl.col('hhvehs') - pl.col('drivers')).alias('auto_count_driver'),
    (pl.col('hhvehs') - pl.col('hhwkrs')).alias('auto_count_worker')
])
df_hh = df_hh.with_columns([
    pl.when(pl.col('drivers') <= 0).then(pl.lit("no driver")).
    when(pl.col('hhvehs') <= 0).then(pl.lit("no car")).
    when(pl.col('auto_count_driver') < 0).then(pl.lit("cars fewer than drivers")).
    otherwise(pl.lit("enough cars")).alias('auto_available_driver'),
    pl.when(pl.col('hhwkrs') <= 0).then(pl.lit("no worker")).
    when(pl.col('hhvehs') <= 0).then(pl.lit("no car")).
    when(pl.col('auto_count_worker') < 0).then(pl.lit("cars fewer than workers")).
    otherwise(pl.lit("enough cars")).alias('auto_available_worker')
])

# add person type labels
ptype_cat = {1: "full time worker",
             2: "part time worker",
             3: "non-worker age 65+",
             4: "other non-working adult",
             5: "university student",
             6: "grade school student/child age 16+",
             7: "child age 5-15",
             8: "child age 0-4"}
df_person = df_person.with_columns(
    pl.col('pptyp').replace_strict(ptype_cat, default=None).alias('pptyp_label')
)

mode_cat = {1: "1: walk",
            2: "2: bike",
            3: "3: sov",
            4: "4: hov 2",
            5: "5: hov 3+",
            6: "6: transit",
            8: "8: school bus",
            9: "9: tnc"}
df_trip = df_trip.with_columns(
    pl.col('mode').replace_strict(mode_cat, default=None).alias('mode_label')
)

dpurp_cat = {1: "Work",
              2: "School",
              3: "Escort",
              4: "Personal Business",
              5: "Shop",
              6: "Meal",
              7: "Social"}
df_trip = df_trip.with_columns(
    pl.col('dpurp').replace_strict(dpurp_cat, default=None).alias('dpurp_label')
)

# df_hh = df_hh.join(df_parcel, how="left", left_on='hhno', right_on='parcelid')
df_person = df_person.join(df_hh,
                           how='left', on=['hhno','source']) # get auto ownership from hh data

df_trip = df_trip.join(df_person, how='left', on=['pno','hhno','source'])
```

```python
hb_trip = df_trip.filter(~pl.col('dpurp').is_in([1, 2, 3]))
```

```python
df_plot = (
    hb_trip.group_by(['source', 'mode_label'])
    .agg([
        pl.col('trexpfac').sum().alias('trexpfac'),
        pl.col('trexpfac').count().alias('sample count')
    ])
    .with_columns(
        (pl.col('trexpfac') / pl.col('trexpfac').sum().over(['source'])).alias('percentage')
    )
)

df_plot_pd = df_plot.sort(['source']).to_pandas()

fig = px.bar(df_plot_pd, x="mode_label", y="percentage", color="source",
             barmode="group",hover_data=['sample count'],title="other home-based trip mode")
fig.update_layout(height=400, width=700, font=dict(size=11),
                  xaxis = dict(dtick = 1, categoryorder='category ascending'),
                  yaxis=dict(tickformat=".2%"))
fig.show()
```

### mode choice by segment

```python
def plot_mode_choice(df: pl.DataFrame, grp_var: str, order_list: dict, title_name: str, n_nol: int, height=400, width=800):
    df_plot = (
        df.group_by(['source', grp_var, 'mode_label'])
        .agg([
            pl.col('trexpfac').sum().alias('trexpfac'),
            pl.col('trexpfac').count().alias('sample count')
        ])
        .with_columns(
            (pl.col('trexpfac') / pl.col('trexpfac').sum().over(['source', grp_var])).alias('percentage')
        )
    )

    df_plot_pd = df_plot.sort(['mode_label']).to_pandas()

    fig = px.bar(df_plot_pd,
                 x="percentage", y="mode_label", color="source",barmode="group",
                 facet_col=grp_var, facet_col_wrap=n_nol, orientation='h',
                 hover_data=['sample count'],
                 category_orders=order_list,
                 title="other home-based trip mode choice by " + title_name)
    fig.update_layout(height=height, width=width)
    fig.for_each_annotation(lambda a: a.update(text = a.text.split("=")[-1]))
    fig.for_each_xaxis(lambda a: a.update(tickformat = ".1%"))
    fig.show()
```
