```python
df_workers = person.filter(pl.col('pwtyp') > 0)
```

```python
df = (
    df_workers.filter(pl.col('pwaudist') > 0)
    .with_columns((pl.col('pwaudist') * pl.col('psexpfac')).alias('wt_dist'))
    .group_by(['source'])
    .agg([
        pl.col('wt_dist').sum().alias('wt_dist'),
        pl.col('psexpfac').sum().alias('psexpfac')
    ])
    .with_columns((pl.col('wt_dist') / pl.col('psexpfac')).alias('Mean Distance to Work by Car (pwaudist)'))
)
df.select(['source', 'Mean Distance to Work by Car (pwaudist)'])
```

```python
def plot_work_location(df: pl.DataFrame, var: str, title_cat: str, sub_name: str):
    df_plot = (
        df.filter(pl.col('pwaudist') > 0)
        .with_columns((pl.col('pwaudist') * pl.col('psexpfac')).alias('wt_dist'))
        .group_by(['source', var])
        .agg([
            pl.col('wt_dist').sum().alias('wt_dist'),
            pl.col('psexpfac').sum().alias('psexpfac'),
            pl.col('psexpfac').count().alias('sample count')
        ])
        .with_columns((pl.col('wt_dist') / pl.col('psexpfac')).alias('average_wt_pwaudist'))
    )

    fig = px.bar(df_plot.to_pandas(), x=var, y='average_wt_pwaudist', color="source",
                 barmode="group", hover_data=['sample count'],
                 title="Work Distance by " + title_cat)
    fig.for_each_annotation(lambda a: a.update(text=sub_name + "=<br>" + a.text.split("=")[-1]))
    fig.update_xaxes(title_text=sub_name)
    fig.update_layout(height=400, width=800, font=dict(size=11),
                      yaxis=dict(tickformat=".2f"))
    fig.for_each_yaxis(lambda a: a.update(tickformat=".2f"))
    fig.show()
```

```python
# distance to school bins from workplace_location.csv
df_workers = df_workers.with_columns([
    pl.when(pl.col('pwaudist') < 1).then(pl.lit('0-1'))
      .when(pl.col('pwaudist') < 2).then(pl.lit('1-2'))
      .when(pl.col('pwaudist') < 5).then(pl.lit('2-5'))
      .when(pl.col('pwaudist') < 15).then(pl.lit('5-15'))
      .otherwise(pl.lit('15+')).alias('distance_to_work_bin')
])
max_bin = 60
bin_size = 2
df_workers = df_workers.with_columns(
    pl.when((pl.col('pwaudist') >= 0) & (pl.col('pwaudist') <= max_bin))
      .then(pl.col('pwaudist').truediv(bin_size).floor().mul(bin_size).cast(pl.Int64).cast(pl.Utf8))
      .otherwise(None)
      .alias('d_work_bin_60mi')
)
```

## Distance Bins

```python
def plot_distance(df: pl.DataFrame, group: str, title_name: str):

    df_group = df.filter(pl.col(group).is_not_null())

    df_plot = (
        df_group.filter(pl.col('distance_to_work_bin').is_not_null())
        .group_by(['source', 'distance_to_work_bin'])
        .agg([
            pl.col('psexpfac').sum().alias('psexpfac'),
            pl.col('psexpfac').count().alias('sample count')
        ])
        .with_columns((100 * pl.col('psexpfac') / pl.col('psexpfac').sum().over(['source'])).alias('percentage'))
    )

    fig1 = px.bar(df_plot.to_pandas(), x='distance_to_work_bin', y="percentage", color="source", barmode="group", hover_data=['sample count'],
                  title=title_name)
    fig1.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig1.update_layout(height=400, width=700, font=dict(size=11))
    fig1.show()

    df_plot = (
        df_group.filter(pl.col('d_work_bin_60mi').is_not_null())
        .group_by(['source', 'd_work_bin_60mi'])
        .agg(pl.col('psexpfac').sum().alias('psexpfac'))
        .with_columns((100 * pl.col('psexpfac') / pl.col('psexpfac').sum().over(['source'])).alias('percentage'))
    )

    fig2 = px.line(df_plot.to_pandas(), x='d_work_bin_60mi', y="percentage", color="source",
                   title=title_name)
    fig2.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig2.update_layout(height=400, width=700, font=dict(size=11))
    fig2.show()

# distance to work
plot_distance(df_workers, 'pwtyp', "worker: distance to work (miles)")
```

```python
def plot_group(df: pl.DataFrame, group: str, title_name: str, height: int = 900):
    df_plot = (
        df.group_by(['source', group, 'distance_to_work_bin'])
        .agg([
            pl.col('psexpfac').sum().alias('psexpfac'),
            pl.col('psexpfac').count().alias('sample count')
        ])
        .with_columns((pl.col('psexpfac') / pl.col('psexpfac').sum().over(['source', group])).alias('percentage'))
    )

    fig = px.bar(df_plot.to_pandas(), x="distance_to_work_bin", y="percentage", color="source", barmode="group",
                 facet_col=group, facet_col_wrap=2, hover_data=['sample count'],
                 title="Distance to work by " + title_name)
    fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig.update_layout(height=height, width=700, font=dict(size=11),
                      yaxis=dict(tickformat=".0%"))
    fig.show()

def plot_group_line(df: pl.DataFrame, group: str, title_name: str, height: int = 400):
    df_plot = (
        df.group_by(['source', group, 'd_work_bin_60mi'])
        .agg(pl.col('psexpfac').sum().alias('psexpfac'))
        .with_columns((pl.col('psexpfac') / pl.col('psexpfac').sum().over(['source', group])).alias('percentage'))
    )

    fig2 = px.line(df_plot.to_pandas(), x='d_work_bin_60mi', y="percentage", color="source",
                   facet_col=group, facet_col_wrap=2,
                   title="Distance to work by " + title_name)
    fig2.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig2.update_layout(height=height, width=700, font=dict(size=11),
                       yaxis=dict(tickformat=".0%"))
    fig2.show()
```

```python
# Get District lookup from parcel geographies

df_lehd = util.read_sqlite_db(input_config, summary_config, "SELECT * FROM observed_lehd_district_flows")


df_lehd = df_lehd.filter(pl.col('model_year').cast(pl.Int64) == int(input_config['base_year']))
df_lehd = df_lehd.with_columns(pl.lit('LEHD').alias('source'))
```

```python
dist_name = {1: '1: Suburban Snohomish',
              2: '2: Everett-Lynwood-Edmonds',
              3: '3: North Seattle-Shoreline',
              4: '4: Seattle CBD',
              5: '5: West-South Seattle',
              6: '6: East Side',
              7: '7: Renton-FedWay-Kent',
              8: '8: Tacoma',
              9: '9: Kitsap',
              10: '10: South Pierce'}

df_lehd = df_lehd.with_columns([
    pl.col('home_dist').replace_strict(dist_name, default=None).alias('home_dist_name'),
    pl.col('work_dist').replace_strict(dist_name, default=None).alias('work_dist_name')
])
```

```python
df_model = (
    person.filter(pl.col('source') == 'model')
    .group_by(['District_home', 'District_work'])
    .agg(pl.col('psexpfac').sum().alias('workers_total'))
    .with_columns([
        pl.lit('model').alias('source'),
        pl.col('District_home').cast(pl.Int64),
        pl.col('District_work').cast(pl.Int64)
    ])
    .with_columns([
        pl.col('District_home').replace_strict(dist_name, default=None).alias('home_dist_name'),
        pl.col('District_work').replace_strict(dist_name, default=None).alias('work_dist_name')
    ])
    .select(['source', 'home_dist_name', 'work_dist_name', 'workers_total'])
)

df_lehd = (
    df_lehd.rename({'S000': 'workers_total'})
    .filter((pl.col('home_dist') > 0) & (pl.col('work_dist') > 0))
    .select(['source', 'home_dist_name', 'work_dist_name', pl.col('workers_total').cast(pl.Float64)])
)
```

```python
_df_model = df_model.pivot(values='workers_total', index='home_dist_name', on='work_dist_name', aggregate_function='sum').to_pandas().set_index('home_dist_name')
_df_lehd = df_lehd.pivot(values='workers_total', index='home_dist_name', on='work_dist_name', aggregate_function='sum').to_pandas().set_index('home_dist_name')
```

```python
df = pl.concat([df_lehd, df_model], how='diagonal')
df.head()

var = 'work_dist_name'
df_plot = df.group_by(['source', var]).agg(pl.col('workers_total').sum().alias('workers_total'))
df_plot

fig = px.bar(df_plot.sort('source', descending=True).to_pandas(), x=var, y='workers_total', color="source",
             barmode="group",
             title="Work District Location")
fig.update_layout(height=400, width=800, font=dict(size=11),
                  xaxis=dict(categoryarray=list(dist_name.values())),
                  yaxis=dict(tickformat=".0f"))
fig.show()
```

#### District Distribution: Model - LEHD

```python
import numpy as np
import matplotlib
import matplotlib.cm as cm

df = (_df_model - _df_lehd) / _df_lehd

# Build a symmetric color range around zero, with a safe fallback when all values
# are 0/NaN/inf so TwoSlopeNorm still has a strictly increasing range.
finite_values = df.to_numpy(dtype=float)
finite_values = finite_values[np.isfinite(finite_values)]
abs_max = float(np.max(np.abs(finite_values))) if finite_values.size else 1.0
if abs_max == 0:
    abs_max = 1.0
abs_min = -abs_max

def make_pretty(styler):

    def background_with_norm(s):
        cmap = cm.get_cmap('RdBu_r', 9)
        norm = matplotlib.colors.TwoSlopeNorm(vmin=abs_min, vcenter=0, vmax=abs_max)
        return [
            '' if not np.isfinite(v) else 'background-color: {:s}'.format(matplotlib.colors.to_hex(cmap(norm(v))))
            for v in s.values
        ]

    styler.set_caption("District distribution")
    styler.format("{:.2%}")
    styler.apply(background_with_norm)
    styler.set_properties(**{'color': 'black'})
    return styler

df[list(dist_name.values())].reindex(list(dist_name.values())).style.pipe(make_pretty)
```
