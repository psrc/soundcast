```python
import plotly.express as px
import polars as pl
import os, sys

module_path = os.path.abspath(os.path.join('../../../..')) # or the path to your source code
sys.path.insert(0, module_path)
import util
```

```python
df_tour = util.get_tour_data(summary_config)
```

```python


# ptype_cat = {1: "1: full time worker",
#              2: "2: part time worker",
#              3: "3: non-worker age 65+",
#              4: "4: other non-working adult",
#              5: "5: university student",
#              6: "6: grade school student/child age 16+",
#              7: "7: child age 5-15",
#              8: "8: child age 0-4"}
pdpurp_cat = {1: "Work",
              2: "School",
              3: "Escort",
              4: "Personal Business",
              5: "Shop",
              6: "Meal",
              7: "Social",
              8: "Social",
              9: "Personal Business"}
# df_person = df_person.with_columns(
#     pl.col('pptyp').replace_strict(ptype_cat, default=None).alias('pptyp_label')
# )
# df_person = df_person.with_columns(pl.col('pptyp_label').cast(pl.Categorical))

df_tour = df_tour.with_columns(
    pl.col('pdpurp').replace_strict(pdpurp_cat, default=None).alias('pdpurp_label')
)
df_tour = df_tour.with_columns(pl.col('pdpurp_label').cast(pl.Categorical))

df_tour = df_tour.with_columns(
    (pl.col('tripsh1') + pl.col('tripsh2')).alias('all_stops')
)
df_tour = df_tour.sort(['source'])
# df_tour = df_tour.join(df_person, how='left', on=['pno','hhno','source'])
```

```python
df_plot = (
    df_tour.group_by(['tripsh1', 'source'])
    .agg(pl.col('toexpfac').sum())
    .with_columns(
        (pl.col('toexpfac') / pl.col('toexpfac').sum().over(['source'])).alias('percentage')
    )
    .filter(pl.col('percentage') > 0.001)
)

df_plot_pd = df_plot.sort(['source']).to_pandas()

fig = px.bar(df_plot_pd, x="tripsh1", y="percentage", color="source",
             barmode="group",title="outbound: number of stops (excluding shares < 1%)",
             hover_data=['toexpfac'])
fig.update_layout(height=400, width=700, font=dict(size=11),
                  xaxis = dict(dtick = 1, categoryorder='category ascending'),
                  yaxis=dict(tickformat=".0%"))
fig.show()
```

```python
df_plot = (
    df_tour.group_by(['tripsh2', 'source'])
    .agg(pl.col('toexpfac').sum())
    .with_columns(
        (pl.col('toexpfac') / pl.col('toexpfac').sum().over(['source'])).alias('percentage')
    )
    .filter(pl.col('percentage') > 0.001)
)

df_plot_pd = df_plot.sort(['source']).to_pandas()

fig = px.bar(df_plot_pd, x="tripsh2", y="percentage", color="source",
             barmode="group",title="inbound: number of stops (excluding shares < 1%)",
             hover_data=['toexpfac'])
fig.update_layout(height=400, width=700, font=dict(size=11),
                  xaxis = dict(dtick = 1, categoryorder='category ascending'),
                  yaxis=dict(tickformat=".0%"))
fig.show()
```

```python
df_plot = (
    df_tour.group_by(['all_stops', 'source'])
    .agg(pl.col('toexpfac').sum())
    .with_columns(
        (pl.col('toexpfac') / pl.col('toexpfac').sum().over(['source'])).alias('percentage')
    )
    .filter(pl.col('percentage') > 0.001)
)

df_plot_pd = df_plot.sort(['source']).to_pandas()

fig = px.bar(df_plot_pd, x="all_stops", y="percentage", color="source",
             barmode="group",title="all stops: number of stops (excluding shares < 1%)",
             hover_data=['toexpfac'])
fig.update_layout(height=400, width=700, font=dict(size=11),
                  xaxis = dict(dtick = 1, categoryorder='category ascending'),
                  yaxis=dict(tickformat=".0%"))
fig.show()
```

## number of stops by purpose

```python
df_plot = (
    df_tour.group_by(['tripsh1', 'pdpurp_label', 'source'])
    .agg(pl.col('toexpfac').sum())
    .with_columns(
        (pl.col('toexpfac') / pl.col('toexpfac').sum().over(['pdpurp_label', 'source'])).alias('percentage')
    )
)
df_plot2 = df_plot.filter(
    (pl.col('percentage') > 0.001) & (pl.col('tripsh1') <= 10)
)

df_plot2_pd = df_plot2.sort(['source', 'pdpurp_label']).to_pandas()

fig = px.bar(df_plot2_pd,
             x="tripsh1", y="percentage", color="source",barmode="group",
             facet_col='pdpurp_label', facet_col_wrap=2,
             hover_data=['toexpfac'],
             title="outbound: number of stops by purpose (excluding shares < 1%)")
fig.update_layout(height=800, width=800)
fig.for_each_annotation(lambda a: a.update(text = a.text.split("=")[-1]))
fig.for_each_xaxis(lambda a: a.update(dict(dtick = 1, categoryorder='category ascending')))
fig.show()
```

- outliers

```python
df_plot.filter((pl.col('percentage') > 0.001) & (pl.col('tripsh1') > 10))
```
